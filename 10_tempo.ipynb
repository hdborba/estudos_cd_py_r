```python
tempo = pd.read_csv('/content/drive/MyDrive/data/tempo.csv', sep=";")
tempo
```

```python
tempo.describe()
```

```python
tempo.groupby(['Aparencia']).size()
```

```python
tempo.groupby(['Aparencia']).describe()
```

```python
# ajusta os valores da coluna Aparencia para "sol, nublado e chuva"
tempo['Aparencia'] = tempo['Aparencia'].replace(['menos'], 'chuva')
tempo
```

```python
tempo['Temperatura'].describe()
```

```python
tempo.loc[tempo['Temperatura'] > tempo['Temperatura'].mean(), 'Temperatura'] = tempo['Temperatura'].median()
tempo
```

```python
tempo.loc[tempo['Umidade'] > tempo['Umidade'].mean(), 'Umidade'] = tempo['Umidade'].median()
tempo['Umidade'].fillna(tempo['Umidade'].median(), inplace=True)
tempo
```

```python
tempo.groupby('Vento').size()
```

```python
tempo['Vento'].fillna('Falso', inplace=True)
tempo
```
